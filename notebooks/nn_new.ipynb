```python
import os
import json
import warnings
import numpy as np
import xarray as xr
import proplot as pplt
pplt.rc['reso']='xx-hi'
pplt.rc['figure.dpi'] = 100
warnings.filterwarnings('ignore')
```

```python
with open('/global/cfs/cdirs/m4334/sferrett/monsoon-kernels/scripts/nn_new/configs.json','r',encoding='utf-8') as f:
    CONFIGS = json.load(f)

FILEDIR    = CONFIGS['paths']['filedir']    
MODELDIR   = CONFIGS['paths']['modeldir']
RESULTSDIR = CONFIGS['paths']['resultsdir']
MODELS_CFG = {m['name']: m for m in CONFIGS['models']}

SPLIT      = 'valid'   
PRTHRESH   = 0.01
LANDTHRESH = 0.5
landfrac   = xr.open_dataarray('/global/cfs/cdirs/m4334/sferrett/monsoon-discovery/data/interim/lf.nc')
```

```python
# ---------------------------------------------------------------------
# Denormalization constants for precipitation
# ---------------------------------------------------------------------
stats_path = os.path.join(FILEDIR,'stats.json')
with open(stats_path,'r',encoding='utf-8') as f:
    STATS = json.load(f)

pr_mean = STATS['pr_mean']
pr_std  = STATS['pr_std']

# ---------------------------------------------------------------------
# True precipitation (phys units, lat-lon-time)
# ---------------------------------------------------------------------
true_ds = xr.open_dataset(os.path.join(FILEDIR,f'{SPLIT}.h5'),engine='h5netcdf')
truepr_norm = true_ds['pr']  # normalized log1p

# ensure dims (time, lat, lon)
if set(truepr_norm.dims)=={'time','lat','lon'}:
    truepr_norm = truepr_norm.transpose('time','lat','lon')
else:
    raise RuntimeError(f"Unexpected dims for true 'pr': {truepr_norm.dims}")

log_pr_true = truepr_norm*pr_std+pr_mean
truepr = xr.DataArray(
    np.expm1(log_pr_true.values),
    dims=truepr_norm.dims,
    coords=truepr_norm.coords,
    name='pr'
).transpose('lat','lon','time').load()  # (lat, lon, time)

# ---------------------------------------------------------------------
# Load predicted precip fields from RESULTSDIR
# New convention: nn_<modelname>_<split>_pr.nc
# ---------------------------------------------------------------------
results = {}

for fname in sorted(os.listdir(RESULTSDIR)):
    if not fname.endswith('.nc'):
        continue
    suffix = f'_{SPLIT}_pr.nc'
    if not fname.startswith('nn_') or not fname.endswith(suffix):
        continue

    # strip prefix 'nn_' and suffix '_<split>_pr.nc'
    core = fname[len('nn_'):-len(suffix)]  # this should be <modelname>
    modelname = core

    predpath = os.path.join(RESULTSDIR,fname)
    with xr.open_dataset(predpath,engine='h5netcdf') as ds:
        predpr = ds['pr']
        # ensure (lat, lon, time)
        if set(predpr.dims)=={'lat','lon','time'}:
            predpr = predpr.transpose('lat','lon','time')
        elif set(predpr.dims)=={'time','lat','lon'}:
            predpr = predpr.transpose('lat','lon','time')
        else:
            raise RuntimeError(f"Unexpected dims for predicted 'pr' in {fname}: {predpr.dims}")
        predpr = predpr.load()

    ytrue,ypred = xr.align(truepr,predpr,join='inner')

    modeldesc = MODELS_CFG.get(modelname,{}).get('description',modelname)
    description = modeldesc  # no separate "run" concept anymore

    key = modelname
    results[key] = dict(description=description,ytrue=ytrue,ypred=ypred)

print(f'Found {len(results)} completed models for split="{SPLIT}".')
```

```python
def get_r2(ytrue,ypred,dims=None,landfrac=None,landthresh=LANDTHRESH):
    dims = list(ytrue.dims) if dims is None else dims
    if landfrac is not None:
        r2values = []
        for mask in [landfrac>=landthresh,landfrac<landthresh]:
            ssres = ((ytrue.where(mask)-ypred.where(mask))**2).sum(dim=dims,skipna=True)
            sstot = ((ytrue.where(mask)-ytrue.where(mask).mean(dim=dims,skipna=True))**2).sum(dim=dims,skipna=True)
            r2values.append(1-ssres/sstot)
        return tuple(r2values)
    ssres = ((ytrue-ypred)**2).sum(dim=dims,skipna=True)
    sstot = ((ytrue-ytrue.mean(dim=dims,skipna=True))**2).sum(dim=dims,skipna=True)
    return 1-ssres/sstot
    
def get_confusion_matrix(ytrue,ypred,dims=None,prthresh=PRTHRESH):
    dims = list(ytrue.dims) if dims is None else dims
    ytrueabove = ytrue>prthresh
    ypredabove = ypred>prthresh
    tp = (ypredabove&ytrueabove).sum(dim=dims,skipna=True)
    tn = (~ypredabove&~ytrueabove).sum(dim=dims,skipna=True)
    fp = (ypredabove&~ytrueabove).sum(dim=dims,skipna=True)
    fn = (~ypredabove&ytrueabove).sum(dim=dims,skipna=True)
    if dims==list(ytrue.dims) and 'member' not in ypred.dims:
        return int(tp),int(fp),int(fn),int(tn)
    return tp,fp,fn,tn
```

```python
def plot_density_panels(results,nrows,ncols,globalmin=0,globalmax=70):
    binedges = np.linspace(globalmin,globalmax,71)
    fig,axs  = pplt.subplots(nrows=nrows,ncols=ncols,refwidth=2,share=True)
    axs.format(suptitle='2D Distributions (Validation Set)',xlabel='Observed Precipitation (mm/hr)',ylabel='Predicted Precipitation (mm/hr)')
    for ax,(runname,result) in zip(axs,results.items()):
        description = result['description']
        ytrue,ypred = result['ytrue'],result['ypred']
        r2 = get_r2(ytrue,ypred,dims=None)
        if 'member' in ypred.dims:
            r2mean,r2std,ypredplot = r2.mean('member'),r2.std('member'),ypred.mean('member')
        else:
            r2mean,r2std,ypredplot = r2,None,ypred
        ytrueflat = ytrue.values.ravel()
        ypredflat = ypredplot.values.ravel()
        finite    = np.isfinite(ytrueflat)&np.isfinite(ypredflat)
        hist,_,_ = np.histogram2d(ytrueflat[finite],ypredflat[finite],bins=[binedges,binedges])
        hist = np.ma.masked_where(hist==0,hist)
        im   = ax.pcolormesh(binedges,binedges,hist.T,cmap='ColdHot',vmin=1,vmax=1500,levels=100,extend='max')
        ax.plot([globalmin,globalmax],[globalmin,globalmax],color='k',linewidth=1,linestyle='--')
        text = f'R$^2$ = {r2mean:.3f}  {r2std:.4f}' if r2std is not None else f'R$^2$ = {r2mean:.3f}'
        ax.text(0.02,0.98,text,transform=ax.transAxes,va='top',ha='left')
        ax.format(title=description,xlim=(globalmin,globalmax),xticks=10,ylim=(globalmin, globalmax),yticks=10)
    fig.colorbar(im,loc='r',label='Counts',ticks=200)
    pplt.show()

def plot_confusion_matrix(results,nrows,ncols,prthresh=PRTHRESH):
    fig,axs = pplt.subplots(nrows=nrows,ncols=ncols,refwidth=2,share=True)
    axs.format(suptitle=f'Confusion Matrix (Threshold = {prthresh} mm/hr)')
    for ax,(runname,result) in zip(axs,results.items()):
        description = result['description']
        ytrue,ypred = result['ytrue'],result['ypred']
        if 'member' in ypred.dims:
            tp,fp,fn,tn = get_confusion_matrix(ytrue,ypred,dims=list(ytrue.dims),prthresh=prthresh)
            total = tp+fp+fn+tn
            tnpct,fppct,fnpct,tppct = [100.0*x/total for x in (tn,fp,fn,tp)]
            matrixmean = np.array([[tnpct.mean('member'),fppct.mean('member')],[fnpct.mean('member'),tppct.mean('member')]],dtype=float)
            matrixstd  = np.array([[tnpct.std('member'),fppct.std('member')],[fnpct.std('member'),tppct.std('member')]],dtype=float)
        else:
            tp,fp,fn,tn = get_confusion_matrix(ytrue,ypred,dims=None,prthresh=prthresh)
            total = tp+fp+fn+tn
            matrixmean = 100.0*np.array([[tn,fp],[fn,tp]],dtype=float)/total
            matrixstd  = np.zeros_like(matrixmean)
        im = ax.pcolormesh([0,1,2],[0,1,2],matrixmean,cmap='Blues',vmin=0,vmax=100,levels=111)
        labels = [('TN',0,0,0.5,0.5),('FP',0,1,1.5,0.5),('FN',1,0,0.5,1.5),('TP',1,1,1.5,1.5)]
        for label,i,j,x,y in labels:
            meanij,stdij = matrixmean[i,j],matrixstd[i,j]
            text = f'{label}\n{meanij:.1f}  {stdij:.1f}%' if np.nanmax(matrixstd)>0 else f'{label}\n{meanij:.1f}%'
            ax.text(x,y,text,ha='center',va='center')
        ax.format(title=description,xlim=(0,2),xticks=[0.5,1.5],xticklabels=['Predicted Dry','Predicted Wet'],xminorticks='none',
                  ylim=(0,2),yticks=[0.5,1.5],yticklabels=['Observed\nDry','Observed\nWet'],yminorticks='none')
    fig.colorbar(im,loc='r',ticks=10,label='Percent of Samples')
    pplt.show()

def plot_r2_maps(results,nrows,ncols,landfrac,latlim=(5,25),lonlim=(60,90)):
    fig,axs = pplt.subplots(nrows=nrows,ncols=ncols,refwidth=3,proj='cyl',share=False)
    axs.format(suptitle='Time-Mean Spatial Distribution of R$^2$',coast=True,latlim=latlim,lonlim=lonlim)
    for ax,(runname,result) in zip(axs,results.items()):
        description = result['description']
        ytrue,ypred = result['ytrue'],result['ypred']
        r2 = get_r2(ytrue,ypred,dims=['time'])
        if 'member' in ypred.dims:
            r2mean,r2std   = r2.mean('member'),r2.std('member')
            r2land,r2ocean = get_r2(ytrue,ypred,dims=['time','lat','lon'],landfrac=landfrac)
            r2landmean,r2landstd   = r2land.mean('member'),r2land.std('member')
            r2oceanmean,r2oceanstd = r2ocean.mean('member'),r2ocean.std('member')
        else:
            r2mean,r2std   = r2,None
            r2land,r2ocean = get_r2(ytrue,ypred,dims=['time','lat','lon'],landfrac=landfrac)
            r2landmean,r2landstd   = r2land,None
            r2oceanmean,r2oceanstd = r2ocean,None
        im = ax.pcolormesh(r2mean.lon,r2mean.lat,r2mean,cmap='ColdHot',cmap_kw={'left':0.5},vmin=0,vmax=0.4,levels=9,extend='both')
        if r2std is not None:
            text = (f'Land R$^2$ = {r2landmean:.3f}  {r2landstd:.3f}\nOcean R$^2$ = {r2oceanmean:.3f}  {r2oceanstd:.3f}')
        else:
            text = (f'Land R$^2$ = {r2landmean:.3f}\nOcean R$^2$ = {r2oceanmean:.3f}')
        ax.text(0.05,0.95,text,transform=ax.transAxes,ha='left',va='top',bbox=dict(boxstyle='round',facecolor='white',alpha=0.8,edgecolor='k'))
        ax.format(title=description)
    fig.colorbar(im,loc='b',label='R$^2$',ticks=0.1)
    pplt.show()

def plot_r2_bars(results):
    labels,r2means,r2stds = [],[],[]
    for runname,result in results.items():
        description = result['description']
        ytrue,ypred = result['ytrue'],result['ypred']
        r2 = get_r2(ytrue,ypred,dims=None)
        if 'member' in ypred.dims:
            r2mean,r2std = r2.mean('member'),r2.std('member')
        else:
            r2mean,r2std = r2,0.0
        labels.append(description)
        r2means.append(r2mean)
        r2stds.append(r2std)
    fig,ax = pplt.subplots(nrows=1,ncols=1,refwidth=4,refheight=2)
    ax.format(suptitle='NN Skill',xlabel='R$^2$',ylabel='')
    ax.barh(labels,r2means,xerr=r2stds,color='red6',capsize=3)
    pplt.show()

def plot_bias_maps(results,nrows,ncols,latlim=(5,25),lonlim=(60,90)):
    biases,labels = [],[]
    for runname,result in results.items():
        description = result['description']
        ytrue,ypred = result['ytrue'],result['ypred']
        if 'member' in ypred.dims:
            ypredmean = ypred.mean('member')
        else:
            ypredmean = ypred
        bias = (ypredmean-ytrue).mean('time')
        biases.append(bias)
        labels.append(description)
    vmax    = max(float(abs(b).max()) for b in biases)
    fig,axs = pplt.subplots(nrows=nrows,ncols=ncols,proj='cyl',refwidth=3,share=False)
    axs.format(suptitle='Time-Mean Precipitation Bias',coast=True,latlim=latlim,latlines=5,lonlim=lonlim,lonlines=5)
    for ax,bias,label in zip(np.atleast_1d(axs),biases,labels):
        im = ax.pcolormesh(bias.lon,bias.lat,bias,cmap='ColdHot',vmin=-0.4,vmax=0.4,levels=21,extend='both')
        ax.format(title=label)
    fig.colorbar(im,loc='b',label='Precipitation Bias (mm/hr)')
    for ax in axs[len(biases):]:
        ax.remove()
    pplt.show()

def plot_bias_doy(models):
    nnruns   = list(results.keys())
    nncolors = pplt.Colormap('ColdHot')(np.linspace(0,1,len(nnruns))) 
    xticks  = [152,166,183,197,214,228,244]
    xlabels = ['Jun 1','Jun 15','Jul 1','Jul 15','Aug 1','Aug 15','Aug 31']
    fig,ax = pplt.subplots(refwidth=6, refheight=2)
    ax.format(suptitle='Daily Climatology of Domain-Mean Precipitation Bias',
              xlabel='Day of Year',xticks=xticks,xticklabels=xlabels,xminorticks='none',
              ylabel='Precipitation Bias (mm/hr)')
    ax.axhline(0,color='k',linestyle='--',linewidth=1)
    for i,(runname,result) in enumerate(results.items()):
        description = result['description']
        ytrue,ypred = result['ytrue'],result['ypred']
        if 'member' in ypred.dims:
            bias = (ypred-ytrue).mean(['lat','lon'])
            clim = bias.groupby('time.dayofyear').mean('time')
            mean,std = clim.mean('member'),clim.std('member')
        else:
            bias = (ypred-ytrue).mean(['lat','lon'])
            mean,std = bias.groupby('time.dayofyear').mean('time'),None
        doy = mean['dayofyear'].values
        if std is not None:
            ax.fill_between(doy,mean.values-std.values,mean.values+std.values,alpha=0.3,color=nncolors[i])
        ax.plot(doy,mean.values,color=nncolors[i],linewidth=2,label=description)
    ax.legend(loc='r',ncols=1)
    pplt.show()
```

```python
# -------------------------------------------------------------------
# Kernel extraction and plotting (vertical kernels only)
# -------------------------------------------------------------------

def extract_vertical_kernel(runname, modelname):
    """
    Load a trained vertical kernel model and return:
    - lev (np.ndarray): pressure levels (hPa)
    - weights (np.ndarray): mean kernel weights over fieldvars (sums to 1)
    """
    # Get config for this run/model
    if runname not in RUNS_CFG:
        raise ValueError(f"Unknown run '{runname}' in configs.json")
    if modelname not in MODELS_CFG:
        raise ValueError(f"Unknown model '{modelname}' in configs.json")

    run_cfg   = RUNS_CFG[runname]
    model_cfg = MODELS_CFG[modelname]

    fieldvars  = run_cfg['fieldvars']
    nfieldvars = len(fieldvars)
    nlocalvars = len(CONFIGS['params']['localvars'])

    # Open train split to get lev coordinate and nlevs
    train_path = os.path.join(FILEDIR, 'train.h5')
    ds_train   = xr.open_dataset(train_path, engine='h5netcdf')
    if 'lev' in ds_train.coords:
        lev = ds_train['lev'].values
    else:
        lev = ds_train[fieldvars[0]].coords['lev'].values
    nlevs = len(lev)

    # Build model on CPU
    mtype  = model_cfg['type']
    axis   = model_cfg.get('axis', None)
    kernel = model_cfg.get('kernel', None)

    # Reuse the same logic as train/eval
    if mtype == 'baseline':
        raise ValueError("Baseline model has no explicit kernel to plot.")

    from models import (
        MainNN,
        NonparametricKernelNN,
        GaussianKernel,
        TopHatKernel,
        ParametricKernelNN,
    )

    if mtype == 'nonparametric':
        if axis != 'lev':
            raise ValueError("This plotting helper assumes axis='lev' for nonparametric.")
        npoints = nlevs
        model = NonparametricKernelNN(
            npoints=npoints,
            nfieldvars=nfieldvars,
            nlocalvars=nlocalvars
        )
    elif mtype == 'parametric':
        if axis != 'lev':
            raise ValueError("This plotting helper assumes axis='lev' for parametric.")
        npoints = nlevs

        def make_layer():
            if kernel == 'gaussian':
                return GaussianKernel
            if kernel == 'tophat':
                return TopHatKernel
            raise ValueError(f"Unknown kernel type '{kernel}'")

        layers = [make_layer()(npoints) for _ in range(nfieldvars)]
        model = ParametricKernelNN(
            kernellayers=layers,
            nlocalvars=nlocalvars
        )
    else:
        raise ValueError(f"Unsupported model type '{mtype}' for kernel plotting.")

    # Load checkpoint
    ckpt_path = os.path.join(MODELDIR, f"{runname}__{modelname}.pt")
    if not os.path.exists(ckpt_path):
        raise FileNotFoundError(f"No checkpoint found at {ckpt_path}")
    state = torch.load(ckpt_path, map_location='cpu')
    model.load_state_dict(state)
    model.eval()

    # Extract kernel weights
    weights_per_field = []

    if mtype == 'nonparametric':
        # model.kernels: list of NonparametricKernel
        for k in model.kernels:
            logits = k.logits.detach().cpu()
            w = torch.softmax(logits, dim=0).numpy()  # (npoints,)
            weights_per_field.append(w)
    else:
        # parametric: model.kernellayers: [GaussianKernel/TopHatKernel,...]
        for k in model.kernellayers:
            npoints = k.npoints
            positions = torch.arange(npoints, dtype=torch.float32)

            if isinstance(k, GaussianKernel):
                center = k.center.detach()
                width  = torch.exp(k.logwidth.detach()) + 1e-6
                dist   = positions - center
                logits = -dist**2 / (2.0 * width**2)
            elif isinstance(k, TopHatKernel):
                center    = k.center.detach()
                halfwidth = torch.exp(k.loghalfwidth.detach()) + 1e-6
                dist      = torch.abs(positions - center)
                inside    = (dist <= halfwidth).float()
                logits    = inside * 0.0 + (1.0 - inside) * (-1e6)
            else:
                raise TypeError(f"Unexpected kernel layer type: {type(k)}")

            w = torch.softmax(logits, dim=0).numpy()
            weights_per_field.append(w)

    weights_per_field = np.stack(weights_per_field, axis=0)  # (nfieldvars, npoints)
    weights = weights_per_field.mean(axis=0)  # average over fields

    # Make sure lev and weights have same length (nlevs)
    if weights.shape[0] != nlevs:
        n = min(weights.shape[0], nlevs)
        weights = weights[:n]
        lev     = lev[:n]

    return lev, weights


def plot_kernels(results, ncols=3):
    """
    Plot learned vertical kernels for all *non-baseline* models that exist
    in 'results' (i.e., have successful predictions saved).

    results: dict from earlier (keys like 'runname__modelname').
    """
    # filter to non-baseline models only
    kernel_entries = []
    for key, result in results.items():
        runname, modelname = key.split('__', 1)
        mtype = MODELS_CFG.get(modelname, {}).get('type', None)
        if mtype in ('nonparametric', 'parametric'):
            kernel_entries.append((runname, modelname, result['description']))

    if not kernel_entries:
        print("No non-baseline kernel models found in results.")
        return

    nplots = len(kernel_entries)
    nrows = int(np.ceil(nplots / ncols))

    fig, axs = pplt.subplots(nrows=nrows, ncols=ncols, refwidth=2, share=False)
    axs = np.atleast_1d(axs)

    for ax, (runname, modelname, description) in zip(axs, kernel_entries):
        try:
            lev, w = extract_vertical_kernel(runname, modelname)
        except Exception as e:
            ax.text(0.5, 0.5, f"Error:\n{e}", ha='center', va='center')
            ax.format(title=f"{runname}__{modelname}")
            continue

        # Plot weights vs pressure
        ax.plot(w, lev, linewidth=2)
        ax.format(
            title=description,
            xlabel='Kernel weight',
            ylabel='Pressure (hPa)',
            yreverse=True  # higher pressure (lower altitude) at bottom
        )

    # Remove unused axes if any
    for ax in axs[nplots:]:
        ax.remove()

    fig.format(suptitle='Learned Vertical Kernels')
    pplt.show()
```
