```python
import os
import json
import warnings
import numpy as np
import pandas as pd
import xarray as xr
from numba import njit
import proplot as pplt
from matplotlib.patches import Patch
warnings.filterwarnings('ignore')
pplt.rc.update({
    'tick.minor':False,
    'savefig.dpi':900,
    'font.size':9,
    'label.size':9,
    'tick.labelsize':9,
    'title.size':9,
    'legend.fontsize':9})

pplt.rc.update({
    'figure.facecolor': 'black',
    'savefig.facecolor': 'black',
    'axes.facecolor': 'white',
    'axes.edgecolor': 'white',
    'text.color':'black',

    # ticks (inside panels)
    'xtick.color': 'white',
    'ytick.color': 'white',
    'xtick.labelcolor': 'white',
    'ytick.labelcolor': 'white',

    # axis labels (outside panels)
    'axes.labelcolor': 'white',

    # titles
    'axes.titlecolor': 'white',
    'suptitle.color': 'white',

    # row/col labels
    'leftlabel.color': 'white',
    'toplabel.color': 'white',
    'rightlabel.color': 'white',
    'bottomlabel.color': 'white',

    # legend
    'legend.facecolor': 'white',    # box background (optional)
    'legend.edgecolor': 'black',    # box edge (optional)
})
```

```python
with open('../scripts/configs.json','r',encoding='utf-8') as f:
    CONFIGS = json.load(f)
SPLITSDIR  = CONFIGS['filepaths']['splits']    
MODELSDIR  = CONFIGS['filepaths']['models']
PREDSDIR   = CONFIGS['filepaths']['predictions']
MODELS     = CONFIGS['models']
SPLIT      = 'test'   
```

```python
@njit
def get_r2(ytrue,ypred):
    ssres = 0.
    ytsum = 0.
    count = 0
    for i in range(ytrue.size):
        yt = ytrue.flat[i]
        yp = ypred.flat[i]
        if not np.isnan(yt):
            ytsum += yt
            count += 1
            if not np.isnan(yp):
                ssres += (yt-yp)**2
    ytmean = ytsum/count
    sstot = 0.
    for i in range(ytrue.size):
        yt = ytrue.flat[i]
        if not np.isnan(yt):
            sstot += (yt-ytmean)**2
    return 1-ssres/sstot

@njit
def get_mse(ytrue,ypred):
    ssres = 0.
    count = 0
    for i in range(ytrue.size):
        yt = ytrue.flat[i]
        yp = ypred.flat[i]
        if not np.isnan(yt) and not np.isnan(yp):
            ssres += (yt-yp)**2
            count += 1
    return ssres/count

@njit
def seed_uncertainty(ytrue,ypred,metric):
    nseed  = ypred.shape[0]
    values = np.empty(nseed)
    for seed in range(nseed):
        values[seed] = metric(ytrue,ypred[seed])
    return values

@njit
def bootstrap_uncertainty(ytrue,ypredmean,metric,niterations=1000):
    nsamples = ytrue.shape[0]
    values   = np.empty(niterations)
    for b in range(niterations):
        idx = np.random.randint(0,nsamples,nsamples)
        values[b] = metric(ytrue[idx,:,:],ypredmean[idx,:,:])
    return values
```

```python
with xr.open_dataset(os.path.join(SPLITSDIR,f'{SPLIT}.h5'),engine='h5netcdf') as ds:
    truepr = ds.pr.load()

rows = []
for model in MODELS:
    name,desc = model['name'],model['description']
    filepath = os.path.join(PREDSDIR,f'{name}_{SPLIT}_predictions.nc')
    if not os.path.exists(filepath):
        continue
    with xr.open_dataset(filepath,engine='h5netcdf') as ds:
        predpr = ds.pr.load()
    ytrue,ypred = xr.align(truepr,predpr,join='inner')
    ytrue = ytrue.transpose('time','lat','lon').values
    ypred = ypred.transpose('seed','time','lat','lon').values
    ypredmean = np.nanmean(ypred,axis=0)
    r2mean  = get_r2(ytrue,ypredmean)
    msemean = get_mse(ytrue,ypredmean)
    # r2seedvalues  = seed_uncertainty(ytrue,ypred,get_r2)
    # mseseedvalues = seed_uncertainty(ytrue,ypred,get_mse)
    # r2bootvalues  = bootstrap_uncertainty(ytrue,ypredmean,get_r2)
    # msebootvalues = bootstrap_uncertainty(ytrue,ypredmean,get_mse)
    # r2seedstd  = r2seedvalues.std()
    # mseseedstd = mseseedvalues.std()
    # r2bootstd  = r2bootvalues.std()
    # msebootstd = msebootvalues.std()
    color = ('yellow4' if ('kernel' in name and 'nonparametric' in name) else 'red4' if ('kernel' in name) else 'blue4')
    rows.append({
        'model':desc,
        'r2_mean':round(float(r2mean),3),
        # 'r2_boot_std':round(float(r2bootstd),4),
        # 'r2_seed_std':round(float(r2seedstd),4),
        'mse_mean':round(float(msemean),3),
        # 'mse_boot_std':round(float(msebootstd),4),
        # 'mse_seed_std':round(float(mseseedstd),4),
        'color':color})
```

```python
pd.set_option('display.max_colwidth',None)
df = (pd.DataFrame(rows).sort_values('r2_mean').reset_index(drop=True))
df.iloc[:,:-1]
```

```python
labels   = df['model'].to_numpy()
colors   = df['color'].to_numpy()
r2means  = df['r2_mean'].to_numpy()
msemeans = df['mse_mean'].to_numpy()
x = np.arange(len(labels))
```

```python
fig,axs = pplt.subplots(nrows=2,figwidth=6.5,refheight=1.2,tight=True,spanx=True,spany=False,space=0.75)
axs.format(grid=False,xrotation=30,ylim=(0.3,0.7),yticks=[0.35,0.45,0.55,0.65])

for ax,data,ylabel in zip(axs,[r2means,msemeans],['R$^2$','MSE']):
    bars = ax.bar(x,data,color=colors)
    for bar,value in zip(bars,data):
        ax.text(bar.get_x()+bar.get_width()/2,bar.get_height()-0.02,f'{value:.3f}',ha='center',va='top')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.format(ylabel=ylabel)
elements = [
    Patch(facecolor='blue4',label='Baseline'),
    Patch(facecolor='yellow4',label='Nonparametric Kernel'),
    Patch(facecolor='red4',label='Parametric Kernel')]
axs[0].legend(handles=elements,loc='ul',ncols=3,frameon=False)
pplt.show()
fig.save('../figs/fig_2A.jpg')
```
